```python
# =====================================================
# === Klasifikasi Cheetah vs Harimau Menggunakan ResNet50 ===
# =====================================================

# --- 1. Import Library ---
import os
import numpy as np
import matplotlib.pyplot as plt
import math
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from google.colab import files

print("✅ Library berhasil diimport.")

# =====================================================
# --- 2. Upload Gambar Manual ---
# =====================================================

# Folder dataset untuk train
base_train_dir = "dataset"
ceetah_train_dir = os.path.join(base_train_dir, "ceetah")
harimau_train_dir = os.path.join(base_train_dir, "harimau")
os.makedirs(ceetah_train_dir, exist_ok=True)
os.makedirs(harimau_train_dir, exist_ok=True)

print("Upload 50 gambar CEETAH untuk training:")
uploaded = files.upload()
for filename, data in uploaded.items():
    with open(os.path.join(ceetah_train_dir, filename), "wb") as f:
        f.write(data)

print("Upload 50 gambar HARIMAU untuk training:")
uploaded = files.upload()
for filename, data in uploaded.items():
    with open(os.path.join(harimau_train_dir, filename), "wb") as f:
        f.write(data)

print("✅ Semua gambar training sudah diupload.")

# Folder dataset untuk test
base_test_dir = "dataset_test"
ceetah_test_dir = os.path.join(base_test_dir, "ceetah")
harimau_test_dir = os.path.join(base_test_dir, "harimau")
os.makedirs(ceetah_test_dir, exist_ok=True)
os.makedirs(harimau_test_dir, exist_ok=True)

print("Upload gambar CEETAH untuk testing:")
uploaded = files.upload()
for filename, data in uploaded.items():
    with open(os.path.join(ceetah_test_dir, filename), "wb") as f:
        f.write(data)

print("Upload gambar HARIMAU untuk testing:")
uploaded = files.upload()
for filename, data in uploaded.items():
    with open(os.path.join(harimau_test_dir, filename), "wb") as f:
        f.write(data)

print("✅ Semua gambar testing sudah diupload.")

# =====================================================
# --- 3. ImageDataGenerator & Augmentasi ---
# =====================================================
IMG_SIZE = (244, 244)
BATCH_SIZE = 10

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = train_datagen.flow_from_directory(
    base_train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    base_train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)

# Test generator
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    base_test_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# =====================================================
# --- 4. Build Model ResNet50 ---
# =====================================================
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(244, 244, 3))

# Bekukan semua layer pretrained
for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(1, activation='sigmoid')(x)  # Biner: Cheetah=0, Harimau=1

model = Model(inputs=base_model.input, outputs=x)

# Kompilasi model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("✅ Model siap untuk dilatih.")
model.summary()

# =====================================================
# --- 5. Training Model ---
# =====================================================
EPOCHS = 10
history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen
)

# =====================================================
# --- 6. Visualisasi Accuracy & Loss ---
# =====================================================
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(1, len(train_acc) + 1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, train_acc, 'bo-', label='Train Accuracy')
plt.plot(epochs_range, val_acc, 'r^-', label='Validation Accuracy')
plt.title('Training & Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, train_loss, 'bo-', label='Train Loss')
plt.plot(epochs_range, val_loss, 'r^-', label='Validation Loss')
plt.title('Training & Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# =====================================================
# --- 7. Evaluasi Model ---
# =====================================================
test_loss, test_acc = model.evaluate(test_generator)
print(f"\nTest Accuracy: {test_acc:.4f}")
print(f"Test Loss: {test_loss:.4f}")

# =====================================================
# --- 8. Simpan & Load Model ---
# =====================================================
model_name = 'cheetah_harimau_resnet.h5'
model.save(model_name, save_format='h5')
loaded_model = tf.keras.models.load_model(model_name)
print(f"✅ Model tersimpan dan berhasil dimuat: {model_name}")

# =====================================================
# --- 9. Prediksi & Visualisasi Test ---
# =====================================================
# Ambil semua gambar test
test_images_all, test_labels_all = [], []
for _ in range(len(test_generator)):
    images, labels = next(test_generator)
    test_images_all.extend(images)
    test_labels_all.extend(labels)

test_images_all = np.array(test_images_all)
test_labels_all = np.array(test_labels_all)

# Prediksi
predictions_all = loaded_model.predict(test_images_all)
predicted_classes_all = (predictions_all > 0.5).astype(int).flatten()

# Visualisasi 25 gambar pertama
numbers_to_display = min(25, len(test_images_all))
num_cells = math.ceil(math.sqrt(numbers_to_display))
plt.figure(figsize=(15, 15))

for plot_index in range(numbers_to_display):
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)

    predicted_label_name = 'Harimau' if predicted_classes_all[plot_index] == 1 else 'Cheetah'
    true_label_name = 'Harimau' if test_labels_all[plot_index] == 1 else 'Cheetah'

    plt.subplot(num_cells, num_cells, plot_index + 1)
    plt.imshow(test_images_all[plot_index])
    plt.xlabel(f"Pred: {predicted_label_name}\nTrue: {true_label_name}", color='black')

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()

# =====================================================
# --- 10. Confusion Matrix ---
# =====================================================
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Hitung confusion matrix
cm = confusion_matrix(test_labels_all, predicted_classes_all)

# Tampilkan dengan heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cheetah', 'Harimau'],
            yticklabels=['Cheetah', 'Harimau'])
plt.xlabel('Prediksi')
plt.ylabel('Label Sebenarnya')
plt.title('Confusion Matrix')
plt.show()

# =====================================================
# --- 11. Classification Report ---
# =====================================================
report = classification_report(test_labels_all, predicted_classes_all,
                               target_names=['Cheetah', 'Harimau'])
print("📊 Classification Report:\n")
print(report)
```
